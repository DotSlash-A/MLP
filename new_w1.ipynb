```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values= np.nan, strategy='mean')
imputer = imputer.fit(data)
data_imputed = imputer.transform(data)
print(data_imputed.shape)
```

```python
imputerr = SimpleImputer(missing_values=np.nan, strategy='mean',add_indicator=True)
imputerr = imputerr.fit(data)
data_impwithindicator=imputerr.transform(data)
print(data_impwithindicator.shape)
```
