```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit

from sklearn.metrics import mean_squared_error

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(306)
shuffle_split_cv = ShuffleSplit(n_splits=10,test_size=0.2,random_state=10)
```

```python
features, labels = fetch_california_housing(as_frame=True, return_X_y=True)
print(features.shape)
assert (features.shape[0]==labels.shape[-1])
labels.shape
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels,test_labels = train_test_split(features,labels,random_state=42)
print(train_features.shape[0])
test_features.shape[0]
```

```python
#pipeline = preprocessing+model
lin_reg_pipeline = Pipeline([("feature_scaling",StandardScaler()),("lin_reg",LinearRegression())])
lin_reg_pipeline.fit(train_features,train_labels) #lin reg with normal eqn
```

```python
print(lin_reg_pipeline[-1].intercept_)#intercept
print(lin_reg_pipeline[-1].coef_)#weight vectors
```

```python
#model evaluation
test_score = lin_reg_pipeline(test_features,test_labels)
print("test score",test_score)
```
